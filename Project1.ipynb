```python
# display daily sales by the week

df.Date=pd.to_datetime(df.Date)
df.dtypes
```

```python
df['Date'].dt.dayofweek
```

```python
day_map={
    0:'Mon',
    1:'Tue',
    2:'Wed',
    3:'Thus',
    4:'Fri',
    5:'Sat',
    6:'Sun'
}
```

```python
df['day_of_week']=df['Date'].dt.dayofweek.map(day_map)
```

```python
df['day_of_week'].value_counts().plot(kind="bar");
```

```python
#What will be the highest months for sales
df['Date'].dt.month
```

```python
month_map={
    1:'Jan',
    2:'Feb',
    3:'Mar',
    4:'Apr',
    5:'May',
    6:'Jun',
    7:'Jul',
    8:'Aug',
    9:'Sep',
    10:'Oct',
    11:'Nov',
    12:'Dec'
}
```

```python
df['Month']=df['Date'].dt.month.map(month_map)
```

```python
df['Month'].value_counts().plot(kind="bar");
```
